```python
import transformers
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
import torch
from datasets import load_dataset
```

```python
dataset = load_dataset('ajibawa-2023/Children-Stories-Collection', split='train[:1%]')
print(dataset.shape)
```

```python
def story_token(example):
    story = example['text'].lower()
    if 'science' in story and 'twist' in story:
      example['text'] = f"[SCIENCE][TWIST] {example['text']}"

    elif 'science' in story:
      example['text'] = f"[SCIENCE] {example['text']}"

    elif 'twist' in story:
      example['text'] = f"[TWIST] {example['text']}"

    elif 'heart' in story:
      example['text'] = f"[HEART] {example['text']}"

    elif 'family' in story:
      example['text'] = f"[FAMILY] {example['text']}"

    elif 'nature' in story:
      example['text'] = f"[NATURE] {example['text']}"

    elif 'love' in story:
      example['text'] = f"[LOVE] {example['text']}"

    elif 'war' in story:
      example['text'] = f"[WAR] {example['text']}"

    elif 'religion' in story:
      example['text'] = f"[RELIGION] {example['text']}"

    elif 'funny' in story:
      example['text'] = f"[FUNNY] {example['text']}"

    elif 'adventure' in story:
      example['text'] = f"[ADVENTURE] {example['text']}"

    else:
      example['text'] = f"[GENERAL] {example['text']}"

    return example


story_dataset = dataset.map(story_token)
print(story_dataset.shape)
```

```python
train_test_split = story_dataset.train_test_split(test_size=0.1)
train_dataset = train_test_split['train']
test_dataset = train_test_split['test']
print(train_dataset.shape)
print(test_dataset.shape)
```

```python
from transformers import GPT2Tokenizer, GPT2LMHeadModel, Trainer, TrainingArguments, DataCollatorForLanguageModeling


model_name = "distilgpt2"
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)
```

```python
print(dataset.column_names)
```

```python
tokenizer.add_special_tokens({"pad_token": "[PAD]"})
story_token = ['[SCIENCE]','[TWIST]','[HEART]', '[FAMILY]','[NATURE]','[LOVE]','[WAR]','[RELIGION]','[FUNNY]','[ADVENTURE]','[GENERAL]']
tokenizer.add_tokens(story_token)
model.resize_token_embeddings(len(tokenizer))
```

```python
def tokenize_function(examples):
   return tokenizer(examples['text'], truncation=True, padding='max_length',max_length=512 )

tokenized_train_datasets = train_dataset.map(tokenize_function, batched=True)
tokenized_test_datasets = test_dataset.map(tokenize_function, batched=True)
```

```python
tokenized_train_dataset = tokenized_train_datasets.remove_columns(['text'])
tokenized_test_dataset = tokenized_test_datasets.remove_columns(['text'])
```

```python
training_args = TrainingArguments(
    output_dir='./story-generator',
    overwrite_output_dir=True,
    num_train_epochs=3,
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    eval_steps=500,
    save_steps=1000,
    warmup_steps=500,
    prediction_loss_only=True,
    logging_dir='./logs',
    logging_steps=100,

)

data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)
trainer = Trainer(
    model = model,
    args = training_args,
    data_collator = data_collator,
    train_dataset = tokenized_train_dataset,
    eval_dataset = tokenized_test_dataset,
)

trainer.train()

trainer.save_model('./story-generator-final')
tokenizer.save_pretrained('./story-generator-final')
```

```python
from transformers import pipeline

# Load the fine-tuned model
story_generator = pipeline('text-generation', model='./story-generator-final', tokenizer='./story-generator-final')

# Generate a story with a specific style
prompt_science = "[SCIENCE] Once upon a time, in a land of talking animals,"
science_story = story_generator(prompt_science, max_length=150, num_return_sequences=1)
print("Generated Science Story:")
print(science_story[0]['generated_text'])

prompt_twist = "[TWIST] A brave knight went on a quest to save a princess, but"
twist_story = story_generator(prompt_twist, max_length=150, num_return_sequences=1)
print("\nGenerated Story with a Twist:")
print(twist_story[0]['generated_text'])
```

```python
import gradio as gr
from transformers import pipeline

# Load the fine-tuned model
story_generator = pipeline('text-generation', model='./story-generator-final', tokenizer='./story-generator-final')

def generate_story(prompt):
    # Generate a story with the given prompt
    generated_text = story_generator(prompt, max_length=150, num_return_sequences=1)[0]['generated_text']
    return generated_text

# Create the Gradio interface
iface = gr.Interface(
    fn=generate_story,
    inputs=gr.Textbox(lines=2, placeholder="Enter your story prompt here..."),
    outputs="text",
    title="Story Generator",
    description="Generate a story based on a given prompt using a fine-tuned GPT-2 model."
)

# Launch the interface
iface.launch(debug=True)
```
